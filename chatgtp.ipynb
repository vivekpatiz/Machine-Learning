```python
b = pd.read_csv('sample_dataset.csv')
```

```python

b['Feedback'] = b['Customer_Feedback'] + " " + b['Website'] + " " + b['Email']
b.head()
```

```python
b.drop(columns=["Customer_Feedback","Website", "Email"], inplace=True)
```

```python
from string import punctuation
b['feeback_puc'] = b['Feedback'].apply(lambda x: ''.join([i for i in x if i not in punctuation]))
```

```python
b.head()
```

```python
from nltk.tokenize import word_tokenize
```

```python
b['feeback_puc'] = b['feeback_puc'].apply( lambda x: word_tokenize(x.lower()))
```

```python
b.head()
```

```python
import nltk
nltk.download('stopwords')
```

```python
stop = nltk.corpus.stopwords.words('english')
stop
```

```python
b['feed_stop'] = b['feeback_puc'].apply(lambda x: [i for i in x if i not in stop])
```

```python
b.head()
```

```python
import re
b['remove'] = b["feed_stop"].apply(lambda x : [ re.sub(r'http/s+|www/s+', '', i ) for i in x ])
```

```python
b.head()
```

```python
ps = nltk.PorterStemmer()
```

```python
b['ps'] = b["remove"].apply(lambda x : [ ps.stem(i) for i in x ])
```

```python
b.head()
```

```python
le = nltk.WordNetLemmatizer()
```

```python
b['le'] = b["ps"].apply(lambda x : [ le.lemmatize(i) for i in x ])
```

```python
b.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
```

```python
b['list_le'] = str(b['le'])
x = cv.fit_transform(b['list_le'])
```

```python
print(x)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tf = TfidfVectorizer()
```

```python
x_X = tf.fit_transform(b['list_le'])
```
